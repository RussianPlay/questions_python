```python
class Company:
    def __init__(self, name, loc):
        self.name = name
        self.loc = loc

    def info(self):
        print(f"Компания: {self.name}, Местоположение: {self.loc}")


class University(Company):
    def __init__(self, name, loc, stud_k):
        super().__init__(name, loc)
        self.stud_k = stud_k

    def add(self, k):
        self.stud_k += k

    def info(self):
        print(f"Университет: {self.name}, Местоположение: {self.loc}, Количество студентов: {self.stud_k}")


c = Company("Кфу", "Казань")
c.info()

u = University("Кфу", "Казань", 20000)
u.info()
```

```python
class Driver:
    def __init__(self, name):
        self.name = name
        self.cars = []

    def add_car(self, car):
        self.cars.append(car)

    def get_name(self):
        print(f"Водитель: {self.name}")

    def info(self):
        print(f"Водитель: {self.name}, Машины:")
        for car in self.cars:
            car.info()


class Car:
    def __init__(self, model, number):
        self.model = model
        self.number = number

    def get_name(self):
        print(f"Модель автомобиля: {self.model}")

    def get_number(self):
        print(f"Номер автомобиля: {self.number}")

    def info(self):
        print(f"Модель автомобиля: {self.model}, Номер автомобиля: {self.number}")


dr = Driver("Эльмир")
c1 = Car("Lamborgini", "12345")
c2 = Car("Ferrari", "00000")
dr.add_car(c1)
dr.add_car(c2)
dr.info()
```

```python
from abc import ABC, abstractmethod


class Game(ABC):
    @abstractmethod
    def play(self):
        pass


class BoardGame(Game):
    def play(self):
        print("Это настольная игра")


class CardGame(Game):
    def play(self):
        print("Это карточная игра")


if __name__ == "__main__":
    board = BoardGame()
    card = CardGame()
    board.play()
    card.play()
```

```python
from abc import ABC, abstractmethod


class IMovie(ABC):
    @abstractmethod
    def updateDirector(self, director: str):
        pass

    @abstractmethod
    def getMovieDetails(self):
        pass


class Movie(IMovie):
    def __init__(self, title, director: str, releaseYear):
        self.title = title
        self.director = director
        self.releaseYear = releaseYear

    def updateDirector(self, director: str):
        self.director = director

    def getMovieDetails(self):
        return f"Фильм: {self.title}, Режиссер: {self.director}, Год выпуска: {self.releaseYear}"


m = Movie("Криминальное чтиво", "Тарантино", 1994)
m.updateDirector("Эльмир")
print(m.getMovieDetails())
```

```python
from abc import ABC, abstractmethod


class University(ABC):
    @abstractmethod
    def add_student(self, student, educ_price):
        pass

    def info(self):
        pass


class PublicUniversity(University):
    def __init__(self):
        self.group = []

    def add_student(self, student, educ_price=0):
        self.group.append((student, educ_price))

    def info(self):
        for student in self.group:
            print(f"Студент: {student[0]}, Оплата: Бесплатно")


class PrivateUniversity(University):
    def __init__(self):
        self.group = []

    def add_student(self, student, educ_price):
        self.group.append((student, educ_price))

    def info(self):
        for student in self.group:
            print(f"Студент: {student[0]}, Оплата: {student[1]}")


def print_university(uni: University):
    uni.info()


if __name__ == "__main__":
    uni1 = PublicUniversity()
    uni1.add_student("Эльмир1")
    uni2 = PrivateUniversity()
    uni2.add_student("Эльмир2", 100000)
    uni2.add_student("Эльмир3", 200000)
    print_university(uni1)
    print_university(uni2)
```

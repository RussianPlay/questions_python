```python
nums_1 = "999912345678900"
nums_2 = "11111111211"
total_num = ""
rem = 0
n = max(len(nums_1), len(nums_2))
nums_1 = nums_1.zfill(n)
nums_2 = nums_2.zfill(n)
for i in range(n - 1, -1, -1):
    num_1, num_2 = int(nums_1[i]), int(nums_2[i])
    total_num += str((num_1 + num_2 + rem) % 10)
    rem = (num_1 + num_2 + rem) // 10
total_num = total_num[::-1]
print(total_num)

```

```python
def find_word_k(t, word):
    lst = t.split()
    print(f"   = {lst.count(word)}")


text = "    ,      "
find_word_k(text, "")
find_word_k(text, "")
find_word_k(text, "")
```

```python
text = ",abc,eeelmir,hi,eeeco,seeeeo."
words = text[:-1].split(",")
k = 0
for i in range(len(words)):
    if words[i].lower().count('e') == 3:
        k += 1
print(k)
```
